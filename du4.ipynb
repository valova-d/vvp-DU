```python
import matplotlib.pyplot as plt
import numpy as np
import math
# R=10, r=3, d=1
# R= 7, r=2, d=5
def graf(R,r,d):
    t = np.linspace(-4* math.pi,4*math.pi,500)
    x = (R-r)* np.cos(t) + d*np.cos(t*(R-r)/r)
    y = (R-r)* np.sin(t) - d*np.sin(t*(R-r)/r)


    fig = plt.figure(figsize=(4,4))
    axes = fig.add_axes((0.1, 0.1, 0.8, 0.8)) 
    graf_funkce = axes.plot(x,y, linewidth=0.5)
    plt.show()

graf(10,3,1)
graf(7,2,5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def slunecnice(n):
    n = np.linspace(1,n, n)
    x = np.sqrt(n)* np.cos(n*math.pi*(3-np.sqrt(5)))
    y = np.sqrt(n)* np.sin(n*math.pi*(3-np.sqrt(5)))
    fig = plt.figure(figsize=(4,4))
    axes = fig.add_axes((0.1, 0.1, 0.8, 0.8)) 
    
    graf_funkce = plt.scatter(x, y, s=8)
    plt.show()

slunecnice(1000)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
def prvocisla(n):
    for i in range (2,n-1):
        if n % i == 0:
            return False
    return True

def prvocisla_graf(n):
    n = np.arange(2,n+1)
    x = np.sqrt(n)* np.cos(n*math.pi*(3-np.sqrt(5)))
    y = np.sqrt(n)* np.sin(n*math.pi*(3-np.sqrt(5)))
    fig = plt.figure(figsize=(4,4))
    axes = fig.add_axes((0.1, 0.1, 0.8, 0.8)) 
    mask = np.array([prvocisla(int(i)) for i in n])
    graf_funkce = plt.scatter(x[mask], y[mask], s=2)
    plt.show()

prvocisla_graf(10000)
```
